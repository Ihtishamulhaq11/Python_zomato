In the phone column the numbers are not in same structure so use 'lambda', 'replace' method to make structure consistent
then fill na and replace Nan values with empty strings.

```python
df['phone']=df['phone'].str.replace('[^a-zA-Z0-9]','')
df['phone']= df['phone'].apply(lambda x:str(x))

df['phone']=df['phone'].replace('^\+91 ','0',regex= True)
df['phone'] = df['phone'].apply(lambda x:x[0:3]+'-'+x[3:7]+'-'+x[7:11])

df['phone']=df['phone'].str.replace('nan--','')
df['phone']=df['phone'].fillna('')
df['phone']=df['phone'].str.replace('nan','')
df['phone'] = df['phone']. str.strip()
```

searched for any inconsistancy in other columns

fill na and replace Nan values with empty strings.

```python
df =df.replace('nan','')
df = df.replace('[]','')
df= df.replace('NaN','',)
df = df.fillna('')
```

changed the data types of column 'rate' and 'approx_cost(for two people)' from object to numerical data type like 'int' and 'float'.
 Renamed the both columns


```python
df['rate'].dtypes
```

```python
df['approx_cost(for two people)'].dtypes
```

```python
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'] ,errors='coerce').astype('Int64')
type(df['approx_cost(for two people)'])
```

```python
df['approx_cost(for two people)'].dtypes
```

```python
df.rename({'approx_cost(for two people)':'apprx_cost(two person)'},axis=1,inplace=True)
```

```python
df['rate'].dtypes
```

```python
df['rate']=df['rate'].str.strip('\/5')
df['rate']  = pd.to_numeric(df['rate']  ,errors='coerce').astype('float')
df['rate'].dtypes
```

```python
df.rename({'rate':'rate_outof_5'},axis=1,inplace=True)
```

Two columns 'dish_liked' and 'cuisines' are in string data type but actually it have list 
So changed the columns type into lists to analyse the data.

```python
df['dish_liked']
```

```python
df['dish_liked']=df['dish_liked'].str.replace(" ","",regex= False)
df['dish_liked'].apply(lambda x: x != '')
df['dish_liked']=df['dish_liked'].str.split(',')
```

```python
df['dish_liked']
```

```python
type(df['dish_liked'][0])
```

```python
df['cuisines']=df['cuisines'].str.replace(" ","",regex= False)
df['cuisines']=df['cuisines'].str.split(',')
```

```python
df['cuisines']
```

```python
df['cuisines'].value_counts()
```

Drop the column 'menu_item' because it have mostly nan and na values

```python
df.menu_item
df = df.drop(columns='menu_item')
```

```python
df_explode = df.explode('cuisines')
stats = df_explode.groupby('cuisines').agg(
    cusine_count = ('cuisines', 'count'),
    median_costs = ('apprx_cost(two person)', 'median')
)
stats=stats.sort_values(by='cusine_count', ascending=False).head(10)
stats
```

```python
stats.plot(kind='scatter', x='cusine_count', y='median_costs')
plt.ylabel('Median cost of cusine')
plt.xlabel('count of customers')
plt.tight_layout()
for i, txt in enumerate(stats.index):
    plt.text(stats['cusine_count'].iloc[i], stats['median_costs'].iloc[i],txt
             )
```
